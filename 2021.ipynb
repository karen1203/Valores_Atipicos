```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
gasto=pd.read_excel('Gasto.xlsx', sheet_name='2021')
gasto
```

Notamos cuantos valores nulos hay dentro del data frame, en cada variable, para después remplazarlos

```python
gasto.isnull().sum()
```

En la siguiente linea se llenó la base de datos con la palabra desconocido, ya que el folio es algo que no se puede reemplazar pues es único.

```python
gasto["FOLIO"]=gasto["FOLIO"].fillna('Desconocido')
gasto
```

En la siguiente línea se bescaba saber cuál era la respuesta más común para remplazar los valores nulos después con el más repetido, que en este caso fue PPD.

```python
f1=gasto.groupby(['MP'])['MP'].count()
f1
```

```python
gasto["MP"]=gasto["MP"].fillna('PPD')
gasto
```

Rectificamos que la limpieza se esté haciendo correctamente con la siguiente función.

```python
gasto.isnull().sum()
```

Dado que la póliza es algo único no se puede repetir, prefiero dejar esos valores nulos como desconocidos.

```python
gasto["POLIZA"]=gasto["POLIZA"].fillna('Desconocida')
gasto
```

Revidso por úlitma vez que ya no haya valores nulos

```python
gasto.isnull().sum()
```

Aquí separamos las variables cualitativas y cuantitativas.

```python
cualitativas=gasto[['FECHA','DESCRIPCION','FOLIO','UUID','RFC','PROVEEDOR', 'TIPO GASTO','TIPO','MP','STATUS']]
cualitativas
```

```python
cuantitativas=gasto[['TC','IMPORTE','IVA','RET ISR','RET IVA', 'TOTAL MX','TOTAL SAT']]
cuantitativas
```

Ahora comenzaré a limpiar los outliers por Rango Intercuartílico ya que me parece más preciso.

Comenzamos limpiando la variable de TC, identificando el primer y tercer cuartil, posterior a eso se calcula el rango intercuartílico y con esos números podemos obtener los límites de los datos y si son mayores que el superior o menores que el inferior los consideramos outliers.

```python
sns.boxplot(data=cuantitativas)
```

```python
y=cuantitativas["TC"]

percentil25=y.quantile(0.25)

percentil75=y.quantile(0.75)

iqr=percentil75-percentil25

limiteSuperior= percentil75+1.5*iqr

limiteInferior=percentil25-1.5*iqr

outliersIqr= cuantitativas[(y>limiteSuperior) | (y<limiteInferior)]

outliersIqr
```

```python
limpio=cuantitativas[(y<=limiteSuperior) & (y>=limiteInferior)]

limpio.shape
limpio
```

Ahora intentamos ver los outliers en la variable de importe.

```python
ye=cuantitativas["IMPORTE"]

per2=ye.quantile(0.25)

per21=ye.quantile(0.75)

iqr1=per21-per2

limSup= per21+1.5*iqr1

limInf=per2-1.5*iqr1

outliersIqr1= cuantitativas[(ye>limSup) | (ye<limInf)]

outliersIqr1
```

Aquí mandamos los outliers a los límites de la variable de Importe de acuerdo al método de rango intercuatílico.

```python
limpio=limpio [(ye<=limSup) & (ye>=limInf)]
limpio.shape
limpio
```

En el siguiente código, al igual que en los anteriores primero buscamos cuáles son los outliers de acuerdo al rango intercuartilico.

```python
a1=cuantitativas["IVA"]

per3=a1.quantile(0.25)

per22=a1.quantile(0.75)

iqr2=per22-per3

limSup1= per22+1.5*iqr2

limInf1=per3-1.5*iqr2

outliersIqr2= cuantitativas[(a1>limSup1) | (a1<limInf1)]

outliersIqr2
```

Y después los acomodamos de acuerdo a los límites

```python
limpio=limpio [(a1<=limSup1) & (a1>=limInf1)]
limpio.shape
limpio
```
