```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
gasto=pd.read_excel('Gasto.xlsx', sheet_name='2023')
gasto
```

Identificamos donde hay valores nulos

```python
gasto.isnull().sum()
```

El folio es único por lo que se deja como desconocido

```python
gasto["Folio"]=gasto["Folio"].fillna('Desconocido')
gasto
```

Identificamos cuál es el más repetido en MP

```python
f1=gasto.groupby(['MP'])['MP'].count()
f1
```

Lo remplazamos con el más repetido

```python
gasto["MP"]=gasto["MP"].fillna('PUE')
gasto
```

Identificamos el valor más repetido

```python
gasto.groupby(['FP'])['FP'].count()
```

Lo reemplazamos por el valor más repetido

```python
gasto["FP"]=gasto["FP"].fillna('99.0')
gasto
```

Volvemos a rectificar cuáles variables tienen valores nulos

```python
gasto.isnull().sum()
```

Vemos cuál es el tipo de gasto más repetido en el data frame y lo reemplazamos por ese

```python
gasto.groupby(['TIPO GASTO'])['TIPO GASTO'].count()
```

Lo reemplazamos por el más repetido que es compras

```python
gasto["TIPO GASTO"]=gasto["TIPO GASTO"].fillna('COMPRAS')
gasto
```

Vemos en que periodos hay más polizas y lo reemplazamos que en este caso es PD02-ENE23



```python
gasto.groupby(['Poliza'])['Poliza'].count()
```

```python
gasto["Poliza"]=gasto["Poliza"].fillna('PD02-ENE23')
gasto
```

Quitamos la variable de otros pues casi todos sus registros están vacios.

```python
gasto=gasto[['Fecha','Folio','UUID','RFC','Proveedor','TIPO GASTO','Descripción','MP','FP','TC','Importe','IVA','RET ISR','RET IVA','TOTAL MX','Tipo', 'Status','Poliza']]
gasto
```

Separamos las variables por cualitativas y cuantitativas, aquí seleccionamos las cualitativas

```python
cualitativos=gasto[['Fecha','Folio','UUID','RFC','Proveedor','TIPO GASTO','Descripción','MP','Tipo','Status','Poliza']]
cualitativos
```

Y aquí seleccionamos las variables cuantitativas

```python
cuantitativos=gasto[['FP','TC','Importe','IVA','RET ISR','RET IVA','TOTAL MX']]
cuantitativos
```

hacemos un boxplor del data frame de las variables cuantitativas para eliminar los outliers.

```python
sns.boxplot(data=cuantitativos)
```

Comenzamos a eliminar los outliers por rangos intercuartílicos

```python
y=cuantitativos["Importe"]

percentil25=y.quantile(0.25)

percentil75=y.quantile(0.75)

iqr=percentil75-percentil25

limiteSuperior= percentil75+1.5*iqr

limiteInferior=percentil25-1.5*iqr

outliersIqr= cuantitativos[(y>limiteSuperior) | (y<limiteInferior)]

outliersIqr
```

```python
cuanti=cuantitativos[(y>limiteSuperior) | (y<limiteInferior)]
```

```python
y1=cuantitativos["IVA"]

percentil251=y1.quantile(0.25)

percentil751=y1.quantile(0.75)

iqr1=percentil751-percentil251

limiteSuperior1= percentil751+1.5*iqr1

limiteInferior1=percentil251-1.5*iqr1

outliersIqr1= cuantitativos[(y1>limiteSuperior1) | (y1<limiteInferior1)]

outliersIqr1
```

```python
cuanti=cuantitativos[(y1>limiteSuperior1) | (y1<limiteInferior1)]
```

```python
cuanti=cuantitativos[(y2>limiteSuperior2) | (y2<limiteInferior2)]
```

```python
y3=cuantitativos["RET IVA"]

percentil253=y3.quantile(0.25)

percentil753=y3.quantile(0.75)

iqr3=percentil753-percentil253

limiteSuperior3= percentil753+1.5*iqr3

limiteInferior3=percentil253-1.5*iqr3

outliersIqr3= cuantitativos[(y3>limiteSuperior3) | (y3<limiteInferior3)]

outliersIqr3
```

```python
y4=cuantitativos["TOTAL MX"]

percentil254=y4.quantile(0.25)

percentil754=y4.quantile(0.75)

iqr4=percentil754-percentil254

limiteSuperior4= percentil754+1.5*iqr4

limiteInferior4=percentil253-1.5*iqr4

outliersIqr4= cuantitativos[(y4>limiteSuperior4) | (y4<limiteInferior4)]

outliersIqr4
```
