```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
gasto=pd.read_excel('Gasto.xlsx', sheet_name='2023')
gasto
```

Identificamos donde hay valores nulos

```python
gasto.isnull().sum()
```

El folio es único por lo que se deja como desconocido

```python
gasto["Folio"]=gasto["Folio"].fillna('Desconocido')
gasto
```

Identificamos cuál es el más repetido en MP

```python
f1=gasto.groupby(['MP'])['MP'].count()
f1
```

Lo remplazamos con el más repetido

```python
gasto["MP"]=gasto["MP"].fillna('PUE')
gasto
```

Identificamos el valor más repetido

```python
gasto.groupby(['FP'])['FP'].count()
```

Lo reemplazamos por el valor más repetido

```python
gasto["FP"]=gasto["FP"].fillna('99.0')
gasto
```

```python
gasto.groupby(['TIPO GASTO'])['TIPO GASTO'].count()
```

```python
gasto["TIPO GASTO"]=gasto["TIPO GASTO"].fillna('COMPRAS')
gasto
```
